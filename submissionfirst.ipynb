```python
import pandas as pd 
train = pd.read_csv("/kaggle/input/titanic/train.csv")
train.head()
```

```python
train["Survived"].value_counts()
```

```python
test= pd.read_csv("/kaggle/input/titanic/test.csv")
```

```python
train["Age"] = train["Age"].fillna(train["Age"].median())
test["Age"]  = test["Age"].fillna(test["Age"].median())

# Clean and encode Sex properly
train["Sex"] = train["Sex"].str.lower().map({"male": 0, "female": 1})
test["Sex"]  = test["Sex"].str.lower().map({"male": 0, "female": 1})

# Safety fill (should not change values, but avoids NaNs)
train["Sex"] = train["Sex"].fillna(0)
test["Sex"]  = test["Sex"].fillna(0)

```

```python
features =["Pclass","Sex","Age","Fare"]

X= train[features]
y= train["Survived"]

X_test= test[features]
```

```python
from sklearn.linear_model import LogisticRegression 

model = LogisticRegression(max_iter=100)
model.fit(X,y)
```

```python
X_test.loc[:, "Fare"] = X_test["Fare"].fillna(X_test["Fare"].median())
```

```python
predictions = model.predict(X_test)
```

```python
X_test.isna().sum()
```

```python
submission = pd.DataFrame({
    "PassengerId": test["PassengerId"],
    "Survived": predictions
})

submission.to_csv("submission.csv", index=False)
submission.head()
```
